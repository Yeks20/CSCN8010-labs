# Linear regression for Median income vs median house value

```python
from sklearn.linear_model import LinearRegression

regression_income = LinearRegression()
regression_income.fit(data[['median_income']], data['median_house_value'])
print('Linear regression: Median House Value vs. Median Income')
print('Intercept:', regression_income.intercept_)
print('Coefficient:', regression_income.coef_[0])
print('Linear regression: Median House Value vs. Median Income')
```

# Linear Regression of Population vs Median House Value

```python
from sklearn.linear_model import LinearRegression

regression_population = LinearRegression()
regression_population.fit(data[['population']], data['median_house_value'])
print('Linear regression: Median House Value vs. Population')
print('Intercept:', regression_population.intercept_)
print('Coefficient:', regression_population.coef_[0])
print()
```

# Linear Regression of Number of Households vs Median House Value

```python
from sklearn.linear_model import LinearRegression

regression_households = LinearRegression()
regression_households.fit(data[['households']], data['median_house_value'])
print('Linear regression: Median House Value vs. Number of Households')
print('Intercept:', regression_households.intercept_)
print('Coefficient:', regression_households.coef_[0])
print()
```
